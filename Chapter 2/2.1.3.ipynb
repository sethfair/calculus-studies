```python
from sympy import *
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sympy.plotting import plot_parametric
%matplotlib inline
```

```python
x = Symbol("x")
```

```python
y = 1 / (1 - x)
y
```

```python
display(x, y)
```

```python
def calculateSlope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)
```

```python
def drawChart(x2):
    y2 = 1 / (1 - x2)
    x1 = 2
    y1 = -1
    xx = np.linspace(2, 4, 100)
    yy = [y.subs('x', thisX) for thisX in xx]
    fig, ax = plt.subplots()
    ax.plot(xx, yy, color="black", linewidth=1)
    ax.grid(linestyle=":")
    slope = calculateSlope(x1, y1, x2, y2)
    ax.set_title("For X: " + str(x2) + ", Slope: " + "{:.2f}".format(slope))
    ax.plot([x1, x2], [y1, y2], color="blue")
    ax.plot(x1, y1, color="red", marker="o", markersize=6)
    ax.plot(x2, y2, color="red", marker="o", markersize=6)
    plt.show()
```

```python
drawChart(1.5)
```

```python
drawChart(1.9)
```

```python
drawChart(1.99)
```

```python
drawChart(1.999)
```

```python
drawChart(2.5)
```

```python
drawChart(2.1)
```

```python
drawChart(2.01)
```

```python
drawChart(2.001)
```

```python
fig, ax = plt.subplots()

ax.set(title='Tangent line for (2, -1)')
ax.grid(linestyle=":")
xx = np.linspace(2, 4, 100)
yy = [y.subs('x', thisX) for thisX in xx]
yy_tangent = [thisX - 3 for thisX in xx]
ax.plot(xx, yy, color="black", linewidth=1)
ax.plot(xx, yy_tangent, 'C1--', linewidth = 2)
ax.plot(2, -1, color="red", marker="o", markersize=6)
plt.show()
```
