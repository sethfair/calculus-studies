```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = (5, 10, 15, 20, 25, 30)
v = (694, 444, 250, 111, 28, 0)
```

```python
slopeList = []
```

```python
def calculateSlope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)
```

```python
def plotSlope(ax, pt1, pt2):
    x1, y1 = pt1
    x2, y2 = pt2
    
    ax.plot(x1, y1, color="red", marker="o", markersize=6)
    ax.plot(x2, y2, color="red", marker="o", markersize=6)
    ax.plot([x1, x2], [y1, y2], color="blue")
    
    slope = calculateSlope(x1, y1, x2, y2)
    slopeList.append(slope)
    plt.text(15, 500, "Slope: " + "{:.2f}".format(slope))
```

```python
markers = np.array([[5, 694], [10, 444], [20, 111], [25, 28], [30, 0]])
markers
```

```python
# show all charts with slope

if slopeList is not []:
    slopeList = []

for x in markers:
    fig, ax = plt.subplots()
    ax.plot(t, v, color="black", linewidth=1)
    ax.set(xlabel='t(min)', ylabel='V(gal)', title='Water tank volume')
    plotSlope(ax, [15, 250], np.asarray(x))
    ax.grid(linestyle=":")
    plt.show()
```

```python
display(mList)
mList = np.array([slopeList[1], slopeList[2]])
display(mList)
slopeAvg = mList.mean()
slopeAvg
```
